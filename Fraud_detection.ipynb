## importing the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
cc_data = pd.read_csv('/content/creditcard.csv')
```

```python
cc_data.head()
```

```python
#time shows the relapsed time of a transaction so 4th is 2 minutes since the first transaction
#all these v1 v2 these are numerical values that are the features that are converted through pricipal component analysis
#class is the label which tells if a transaction is fraudent or normal transaction
```

```python
cc_data.tail()
```

```python
# we have about 2 lakh data points
# this is unbalanced we will do some pre processing , first we get info about the data set
cc_data.info()
# here we dont have any null values the data is clean
```

```python
#another method you already know is by the method is null
cc_data.isnull().sum()
```

```python
#now seeing the legit and fraudlent transactions
cc_data['Class'].value_counts()
```

# this data is highly unbalanced , ml model cannot be trained on this since one label is too large which can import/introduce biasness

0 >> normal transaction

1>> fraudlent transaction

```python
# now we seperate the data for analysis
legit = cc_data[cc_data.Class == 0] # we made a variable and stored the values of one category in that label
fraud = cc_data[cc_data.Class == 1]

```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#some measures that we need about the data
legit.Amount.describe() #take the amount column only for the legit category class 0
```

```python
fraud.Amount.describe()
```

```python
#comparing the value between both
cc_data.groupby('Class').mean()
```

## now will be be doing undersampling

```python
#so building  a  sample dataset contiaining simila distribution of normal transaction and fraudlent transactions
```

number of fraudlent transaction = 492


```python
#what we can do is take some of the data from the legit data set to make it uniform
```

```python
legit_sample = legit.sample(n=492)
```

```python
# now we will concatinate both these sample legit and fraudent, we have a better dataframe balanced one
new_dataset = pd.concat([legit_sample, fraud], axis = 0) # we are adding the value row wise so thats why we used axis = 0
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
#splitting the data into features and targets(0 or 1)
x = new_dataset.drop(columns = 'Class', axis = 1)
y = new_dataset['Class']
```

```python
print(x)
```

```python
print(y)
```

```python
#now splitting the data into training and testing data
x_train, x_test , y_train, y_test = train_test_split(x, y , test_size =0.2, stratify = y, random_state = 2)

#test size is 20 percent of data goes to test data , we use stratify to evenly distribute the data , randomstate just tells the manner at which the data will be splitted
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

# model training


```python
model = LogisticRegression()
```

```python
#here its binary classification so we will be using logistic and its generally used , we can use others as well and try what is getting more accuracy
```

```python
model.fit(x_train, y_train)
```

```python
model = LogisticRegression(max_iter=200)  # Increase max_iter to 200 or higher if needed
model.fit(x_train, y_train)
```

```python
# model evaluation
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
print('accuracy on traning data:', training_data_accuracy)
```

```python
#now we will check for overfitting by seeing the accuracy on test data
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
print('accuracy on test data:', test_data_accuracy)
```

```python
#as we can see the model performed well on the test data as well , we can conclude that the training was succesful and theres no overfitting
```

```python
for i in range(6):  # Adjust range for more examples
    print("Example", i+1)
    print("Actual:", y_test.iloc[i])
    print("Predicted:", model.predict([x_test.iloc[i]])[0])
    print("Features:", x_test.iloc[i].values)
    print()
```

```python
feature_importance = pd.DataFrame(model.coef_.flatten(), x.columns, columns=['Importance'])
print(feature_importance.sort_values(by='Importance', ascending=False))
```

```python
x_original = cc_data.drop(columns='Class', axis=1)
y_original = cc_data['Class']

# Make predictions on the original dataset
y_pred_original = model.predict(x_original)

# Evaluate using accuracy, precision, recall, and F1-score
from sklearn.metrics import classification_report
print(classification_report(y_original, y_pred_original, target_names=["Legit", "Fraud"]))
```
